### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

**New Column of Fares in Euros**

```python
conversion_rate = 0.86
fare_euros_df = dataFrame.copy()
fare_euros_df["Fare Euros"] = fare_euros_df["Fare"] * conversion_rate
fare_euros_df
```

**Subtract Mean Age from Each Age value**

```python
mean_age = dataFrame["Age"].mean()
subtract_mean_age_df = dataFrame.copy()
subtract_mean_age_df["Age"] = subtract_mean_age_df["Age"] - mean_age
subtract_mean_age_df
```

**Subtraction of Custom Array From 2 Columns**

```python
random_numbers = np.random.randint(low=0, high=10, size=len(dataFrame))
selected_num_columns = ["Pclass", "Fare"]
subtracted_df = dataFrame.copy()
subtracted_df[selected_num_columns] = subtracted_df[selected_num_columns].sub(random_numbers, axis=0)
subtracted_df
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



**Convert Genders to Numeric Codes**

```python
num_genders_df = dataFrame["Sex"].map(lambda i: 0 if i == "male" else 1)
num_genders_df
```

**Age Categorizes Column**

```python
def categorize_age(x):
    if x < 18:
        return "child"
    elif x >= 18 and x < 50:
        return "adult"
    else:
        return "senior"

age_categories_df = dataFrame["Age"].map(categorize_age)
age_categories_df
```
