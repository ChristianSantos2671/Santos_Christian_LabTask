### Part 8 ###
Selected **Part 2.1, Part 3.1, and Part 5.1** to redo in pure Python code

```python
# Import these two standard library modules. They are both part of Python
# itself and are not external libraries.  For these, "pip install" is not required.
import urllib.request
import csv

url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"

with urllib.request.urlopen(url) as response:
    lines = [line.decode('utf-8') for line in response]

dataset_list = list(csv.reader(lines))
# The first element of dataset_list is the header row, and the rest are data rows.
```

**Select "Name" Column**

Recoding of Part 2.1.
- Get the header row.
- Find the index in this row of the header "Name".
- Get the values from the subsequent data rows corresponding to that index.

```python
headers = dataset_list[0]
name_index = headers.index("Name")

name_column = []
for row in dataset_list[1:]:    # The 1: starts with the data row and skips header
    name_column.append(row[name_index])

name_column
```

**Older Than 60**

Recoding of Part 3.1.

- Get the header row and find the index of the "Age" column.
- Create an empty list and iterate through all the data rows.
- Append to the list only the rows where the age is above 60.

```python
age_index = headers.index("Age")
older_than_60 = []
for row in dataset_list[1:]:
    age_value = row[age_index]
    if age_value != "" and float(age_value) > 60:
        older_than_60.append(row)

older_than_60
```

**New Column of Fares in Euros**

Recoding of Part 5.1.
- Lookup the conversion rate of Euros and create a conversion factor constant.
- Find the indexs of the "Fare" header.
- Create a copy of the original dataset_list so we modify the copy, not the original.
- Create a new "Fare Euros" header for the new column.
- For every data row, compute the Euro value and append it to that row.
- Display the result.

```python
import copy

conversion_rate = 0.86    #  Euros per USD.
fare_index = headers.index("Fare")

dataset_with_euros = copy.deepcopy(dataset_list) # create a copy of original
dataset_with_euros[0].append("Fare Euros")

# Add the new Euros fare value to every data row.
for row in dataset_with_euros[1:]:
    fare_euros = round(float(row[fare_index]) * conversion_rate, 2)
    row.append(str(fare_euros))

# Display the result.
name_index = headers.index("Name")
euros_index = dataset_with_euros[0].index("Fare Euros")

for row in dataset_with_euros[1:]:
    print(row[name_index], "→", row[fare_index], "USD →", row[euros_index], "EUR")
```
